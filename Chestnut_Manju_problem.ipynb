# Problem: The day when chestnut buns cover the solar system

```python
import math
import matplotlib.pyplot as plt

#write the data we are going to work with
solar_system = (4/3)* math.pi * (4.545 * 1e9) ** 3  #value of the solar system in km^3
solar_system_cm3 = solar_system * 1e15              #value converted to cm^3
chestnut_volume = 100                               #value in cm^3
minutes_dubling = 5                                 #time for the chestnut to multiply

#Create the function for calculate the operation
def calculating_dubling_time(volume_1, volume_2):
  times_doubling = 0                #times the chestnut multiplies
  volumes = [volume_1]              #list for volumes
  minutes = [0]                     #list for time in minutes
  while volume_1 <= volume_2:
    volume_1 *= 2                   #we double the volume
    times_doubling += 1             #we add the times in 1
    volumes.append(volume_1)        #add the new volume in the volume list
    minutes.append(times_doubling * minutes_dubling)  #add the new time in the minute list
  return times_doubling, volumes, minutes

#We call the function with the arguments we previously created
times_doubling, volumes, minutes = calculating_dubling_time(chestnut_volume, solar_system_cm3)
total_minutes_dubling = times_doubling * minutes_dubling          #calculate total minutes

print("It will take approximately {} minutes to reach the target volume.".format(total_minutes_dubling))

#Graphical representation
#We customize the graphic according to our taste
plt.figure(figsize=(8, 5))
plt.plot(minutes, volumes ,color="red", linestyle="-", linewidth=2, marker= 'x',label="Volume of the Chestnut")
plt.axhline(y=solar_system_cm3, color="blue", linestyle="--", linewidth=2, label="Volume of the Solar System")
plt.title("Chestnut Bun Volume Doubling Over Time", fontsize=18)
plt.xlabel("Time(min)")
plt.ylabel("Volume(cm^3)")
plt.legend()
plt.grid(True)
plt.show()
```
